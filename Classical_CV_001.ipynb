```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

%matplotlib notebook
```

```python
#Create a matrix with the size of 50x50, mean of 3, and standard deviation of 7.

def get_gaussian_filter(size, sigma, mean=0):
    ax = np.linspace(-(size - 1) / 2.0, (size - 1) / 2.0, size)
    x_vector, y_vector = np.meshgrid(ax, ax)

    kernel = np.exp(-0.5 * (np.square(x_vector) + np.square(y_vector)) / np.square(sigma))

    return kernel / np.sum(kernel) + mean
```

```python
size = 50
sd = 7
mean = 3

gaussian_filter = get_gaussian_filter(size, sd, mean)

plt.figure()
plt.imshow(gaussian_filter,interpolation='nearest', cmap='gray')
plt.show()
```

```python
image_path = 'games.jpg'
image = cv2.imread(image_path)

plt.figure()
plt.imshow(image)
plt.show()
```

```python
blur = cv2.GaussianBlur(image,(5,5),3)

plt.figure()
plt.imshow(blur)
plt.show()
```

Canny edge detector

```python
edges = cv2.Canny(image,100,200)

plt.figure()
plt.imshow(edges)

edges = cv2.Canny(image,10,20)

plt.figure()
plt.imshow(edges)

edges = cv2.Canny(image,50,300)

plt.figure()
plt.imshow(edges)
plt.show()
```

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

gray_image = np.float32(gray_image)
dst = cv2.cornerHarris(gray_image,2,3,0.04)

dst = cv2.dilate(dst,None)

image_with_edges = np.copy(image)
image_with_edges[dst>0.01*dst.max()]=[0,0,255]

plt.figure()
plt.imshow(image_with_edges)
plt.show()
```
